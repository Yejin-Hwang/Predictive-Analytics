# Predict

```r
#Find the predicted value and a confidence interval

#create a dataframe for the predictor(s)
ndat= data.frame(speed=14)

#To make a prediction when speed = 14
# fit is related to the model with intercept
predict(fit1,newdata=ndat)


#Example : based on the predictive model without the intercept, predict the stopping distance if the speed is 14.
#fit 2 is related to the model without intercept
predict(fit2,newdata=ndat)
```

# Add a confidence/prediction interval 

```r
#To add a confidence interval result
# interval = 'confidence', Level = 0.95

predict(fit1, newdata=ndat, interval='confidence', level = 0.95)

#To add a prediction interval result
# interval = 'prediction', Level = 0.95

predict(fit1, newdata=ndat, interval='predict', level = 0.95)
```

# Residual plot

It can be helpful to check if the model has a good fit or not. In a redsidual plot, we plot the predictor against the residuals. or we plot the predicted values against residuals

```r
# Residual plot
# to obtain the residuals of the model with the intercept

e1 = fit1$residuals

#fitted values of the model
fit.v = fit1$fitted

#cars$speed : 
plot(cars$speed, e1)

#fit.v : predicted values
#e1 : residuals
plot(fit.v,e1)

# If there are any curved patterns, ..., maybe we can improve the model.
```

# Performance measure for a predictive model

## MSE, RMSE, MAD

```r
#the difference between the actual values and the predicted values(fitted values)
rd0=cars$dist - fit1$fitted.values 

#MSE (Mean Squared Error)
MSE=mean(rd0^2)
MSE    

#RMSE
sqrt(MSE)

#MAD
median(abs(rd0))
```

### R-suqared 


**R-squared:** $r^2$ (percentage value)

**where $r$ is the correlation between the predicted values and the true observed values.**

```r
#R-suqared 
r= cor(cars$dist,fit$fitted.values)
RSQ = r^2
RSQ

#About 65% information in the response is explained by the model.
```

# Cross validation

```r
# R code for cross validation
#fit = lm (dist~speed,data=cars)

#Split the data into a training set and testing set

#setting seed to generate a reproducible random sampling

set.seed(1251)

data(cars)
dat=cars

#creating training data as 85% of the dataset
library(caret)
train.index = createDataPartition(dat$Mileage, p = 0.85, list = FALSE)

#generating training dataset from the random_sample
training.dat = dat[train.index,]

#generating testing dataset from rows which are not included in training set

testing.dat= dat[-train.index,]

#str(dat)
#str(training.dat)

#Fit a model using the training data
fit.train = lm(Price~Mileage+Cylinder, dat=training.dat)
```

```r
#Predict the price using the test data (predicted value)
pre.price = predict(fit.train, newdata=testing.dat)

#prediction errors in the testing data
err = testing.dat$Price - pre.price

#Prediction MSE
MSE = mean(err^2)
print(MSE) #     .. >  seed   !


#R squared
r=cor(pre.price,testing.dat$Price)

RSQUARE=r^2

print(RSQUARE)

#According to R^2=.014, the predictive model seems "not good".
```

# Cross validation (LOOCV) 

- trainControl : library(caret)   
- train : library(caret)   
- cv = cross validation   !

```r
# Cross validation R codes
library(caret)

# Load the cars dataset and confirm it's correct
data(cars, package = "datasets")
dat <- as.data.frame(cars)
head(dat)

#Leave-One-Out Cross-Validation (LOOCV)
set.seed(1241)

cv.control = trainControl(method = "LOOCV")

#Linear regression
fit = train( dist ~ speed, data = dat, method ="lm", trControl = cv.control)
summary(fit)

#Fit on the whole data. no difference for 'lm'
fit2 = lm(dist ~ speed, data = dat)
summary(fit2)
```

# k-Fold Cross-Validation 

- library(caret) 
- trainControl(method = "cv", number = k) 

```r
# k-Fold Cross-Validation, k = 8
cv.control = trainControl(method = "cv", number = 8)

# Linear regression
fit.cv = train(dist ~ speed, data = dat, method ="lm", trControl = cv.control)
summary(fit.cv)

# Repeated K-Fold Cross-Validation, k = 10, repeats = 10
set.seed(32421)
cv.control = trainControl(
   method = "repeatedcv",
   number = 10,
   repeats = 10
)

# Linear regression
fit.rcv = train(dist ~ speed, data = dat, method ="lm", trControl = cv.control)
summary(fit.rcv)

# The resampling method options:
# "boot", "boot632", "optimism_boot", "boot_all",
# "cv", "repeatedcv", "LOOCV", "LGOCV" (for repeated training/test splits),
# "none" (only fits one model to the entire training set),
# "oob" (only for random forest, bagged trees, bagged earth, bagged flexible discriminant analysis),
# "timeslice", "adaptive_cv", "adaptive_boot", or "adaptive_LGOCV"
```
